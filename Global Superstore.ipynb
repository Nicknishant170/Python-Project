```python
#Exploratory Data Analysis
sns.countplot(x = 'Order Priority',data = df)
```

```python
#REGION
# Assuming 'Region' is a categorical column in the DataFrame 'df'
region_counts = df['Region'].value_counts()

# Create a pie chart
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Distribution of Regions')

plt.show()
```

```python
# plotting a bar chart for Order Priority and it's count

ax = sns.countplot(x = 'Order Priority',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for Order Priority vs Sales
Sales_gen = df.groupby(['Order Priority'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

sns.barplot(x = 'Order Priority',y= 'Sales' ,data = Sales_gen)
```

```python
ax = sns.countplot(data = df, x = 'Order Priority', hue = 'Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Segment', hue = 'Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Sales')
```

```python
# total amount/sales from top 10 City

sales_state = df.groupby(['City'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(14,8)})
sns.barplot(data = sales_state, x = 'City',y= 'Sales')
```

```python
# SEGMENT

ax = sns.countplot(data = df, x = 'Segment')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#CATEGORY

ax = sns.countplot(data = df, x = 'Category')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#SUBCATEGORY 

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Sub Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#SUBCATEGORY VS SALES 

sales_subcategory = df.groupby(['Sub Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_subcategory, x = 'Sub Category',y= 'Sales')
```

```python
#PRDOUCT ID VS QUANTITY

sales_state = df.groupby(['Product ID'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False).head(10)

sns.set(rc={'figure.figsize':(21,8)})
sns.barplot(data = sales_state, x = 'Product ID',y= 'Quantity')
```

```python
# top 10 most Sales of the products 

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product ID')['Sales'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
