```python
#LIST OF CATEGORICAL VARIABLES TO PLOT
#COUNTPLOT 

#Method 1:

cat_vars=['Gender','Occupation','BMI Category']
#Create Figure with Subplots 

fig, axs = plt.subplots(nrows=1,ncols=3,figsize=(20,8))
axs = axs.flatten()

#Create Barplot for each categorical Variables
for i, var in enumerate(cat_vars):
    sns.countplot(x=var, hue='Sleep Disorder', data=df, ax=axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)
    
#Adjust the spacing between Subplots
fig.tight_layout()
#Show Plt
plt.show()

# enumerate is a convenient way to get both the index and the value of each item in the list during iteration
```

```python
#Method 2:

# #Exploratory Data Analysis
# plt.figure(figsize=(20, 8))
# sns.countplot(x='Occupation',data = df)


# Assuming 'col_list' contains the list of columns you want to plot
col_list = ['Gender', 'Occupation', 'BMI Category']  # Add your desired column names here

plt.figure(figsize=(20, 8))
axs = plt.subplots(1, len(col_list), figsize=(20, 8), squeeze=False)[1].ravel()

for i, col in enumerate(col_list):
    sns.countplot(x=col, hue='Sleep Disorder', data=df, ax=axs[i], dodge=True)
    axs[i].set_title(f'Count plot of {col}', fontsize=15)
    axs[i].tick_params(axis='x', rotation=90)
plt.show()


# # plt.subplots(1, len(col_list), figsize=(20, 8), squeeze=False) creates a grid of subplots where the first argument, 1, 
# # represents the number of rows, len(col_list) represents the number of columns, and figsize=(20, 8) 
# determines the overall size of the entire grid of subplots. The squeeze=False argument ensures that the returned array is always a 2D array.

# # [1] accesses the second element of the tuple returned by plt.subplots(), which is the array of axes 
# # (i.e., the subplot objects).

# # .ravel() is a method that converts the 2D array of subplots into a 1D array. 
# # This is helpful for easier indexing and iteration when creating the subplots.
```

```python
#HISTOGRAM 

#Methdd 1:


import warnings
warnings.filterwarnings('ignore')

#get list of categorical variables
cat_vars=['Gender','Occupation']

#create Figure with subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,6))

#create histplot for each categorical variables
for i, var in enumerate(cat_vars):
    sns.histplot(x=var ,hue='Sleep Disorder', data=df, ax=axs[i],multiple='fill',kde=False, element='bars',fill=True,stat='density')
    axs[i].set_xticklabels(df[var].unique(),rotation=90)
    axs[i].set_xticklabels(var)

#Adjust Spacing between Subplots
fig.tight_layout()

#Show Plot
plt.show()            
```

```python
#Method 2 :

import warnings
warnings.filterwarnings('ignore')
# get list of categorical variables
col_list = ['Gender', 'Occupation', 'BMI Category']  # Add your desired column names here

fig, axs = plt.subplots(1, len(col_list), figsize=(20, 6), sharey=True)

for i, col in enumerate(col_list):
    sns.histplot(x=col, hue='Sleep Disorder', data=df, ax=axs[i], multiple='stack')
    axs[i].set_title(f'Histogram of {col}', fontsize=15)
    axs[i].tick_params(axis='x', rotation=90)

plt.show()


# sns.histplot(x=col, hue='Sleep Disorder', data=df, ax=axs[i], multiple='stack'): 
# This line creates a histogram using Seaborn's histplot function. It plots the data from the DataFrame df for the column 
# specified by col with 'Sleep Disorder' as the hue. The ax=axs[i] parameter specifies that the current subplot axs[i] should 
# be used for this particular plot.

# axs[i].set_title(f'Histogram of {col}', fontsize=15): This line sets the title for the current subplot with the column name as the title.

# axs[i].tick_params(axis='x', rotation=90): This line adjusts the x-axis tick labels' rotation to 90 degrees for better readability.
```

```python
#PIE CHART 

import warnings
warnings.filterwarnings('ignore')
max_categories = 5 # Adjust the maximum number of categories to highlight
# get list of categorical variables
col_list = ['Gender', 'Occupation', 'BMI Category']  # Add your desired column names here

fig, axs = plt.subplots(1, len(col_list), figsize=(24, 14))

# Custom color palettes for each pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

for i, col in enumerate(col_list):
    category_counts = df[col].value_counts()
    other_category_count = category_counts[max_categories:].sum()
    highlighted_categories = category_counts[:max_categories]
    highlighted_categories['Other'] = other_category_count
    axs[i].pie(highlighted_categories, labels=highlighted_categories.index, autopct='%1.1f%%', startangle=90, colors=colors)
    axs[i].set_title(f'{col}', fontsize=18)

plt.show()
```
